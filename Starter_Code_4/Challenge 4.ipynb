```python
import pandas as pd 
from pathlib import Path
import numpy as np
import math

whale_navs_df = pd.read_csv(
    Path("Resources/whale_navs.csv"),
    index_col='date',
    infer_datetime_format=True,
    parse_dates=True
)

whale_navs_df = whale_navs_df.pct_change().dropna()

```

```python
#Cumulative Returns for all funds and S&P

cumulative_returns = whale_navs_df.cumsum()

cumulative_returns.plot(title= "Cumulative Returns", kind='line', figsize =(20,10))
cumulative_returns.head()

#I could not figure out how to get cumprod() to work with this dataframe. It gives values that are basically 0. I used cumsum() instead
print("Based on the plot of cumulative returns, it is clear that none of the firms outperformed the S&P over nearly 6 years")
```

```python
#Measuring volatility using cumulative returns
cumulative_returns.plot(title= "Volatility Measure", kind='box', figsize =(15,10))
cumulative_returns.head()

#Fund performance, dropped S&P.
fund_performance = cumulative_returns.drop('S&P 500',1)

fund_performance.plot(title= "Fund Perfomance", kind='box', figsize = (20,10) )

print("Berkshire Hathaway was the most volatile fund while Tiger Glboal Management was the least volatile")
```

```python
#Annualized Standard Deviation
annualized_std = cumulative_returns.std() * math.sqrt(252)

annualized_std.plot(title= 'Annualized Std Deviation', kind='bar')

#Having trouble figuring out how to add a rolling window of 21 days...
```

```python
#Annualized Standard Deviation for Firms
annualized_std_firms = fund_performance.std() * math.sqrt(252)

annualized_std_firms.plot(title= 'Fund Std Deviation', kind='bar')
```
