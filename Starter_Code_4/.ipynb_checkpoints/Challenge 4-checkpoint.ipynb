## Import Dependencies and Read CSV into Dataframe

```python
import pandas as pd 
from pathlib import Path
import numpy as np
import math

whale_navs_df = pd.read_csv(
    Path("Resources/whale_navs.csv"),
    index_col='date',
    infer_datetime_format=True,
    parse_dates=True
)

whale_navs_df = whale_navs_df.pct_change().dropna()

whale_navs_df

```

## Analyze the Performance

```python
#Cumulative Returns for all funds and S&P

cumulative_returns = (1+whale_navs_df).cumprod()
#cumulative_returns
```

```python
cumulative_returns.plot(title= "Cumulative Returns", kind='line', figsize =(10,7))
cumulative_returns.head()

print("Based on the plot of cumulative returns, it is clear that none of the firms outperformed the S&P over nearly 6 years")
```

Based on the cumulative return data and the visualization, none of the firms outperformed the S&P 500 over time. 
There were times in 2015-2016 where most of the firms outperformed the S&P, but the S&P recovered and overtook the spot for highest cumulative returns.

It is important to note that the S&P increased the greatest volatility in cumulative returns as well. 
While none of the firms outperformed the S&P, they had much less volatility. 

## Analyze the Volatility

```python
#Measuring volatility using cumulative returns
cumulative_returns.plot(title= "Volatility Measure", kind='box', figsize =(10,7))
cumulative_returns.head()

#Fund performance, dropped S&P.
fund_performance = cumulative_returns.drop('S&P 500',1)

fund_performance.plot(title= "Fund Perfomance", kind='box', figsize = (10,7))

print("Berkshire Hathaway was the most volatile fund while Tiger Glboal Management was the least volatile")
```

The most volatile fund was Berkshire Hathaway, with performance ranging from ~1 - ~1.25
Other funds had a much smaller range between .95 and 1.05, with the smallest range being Tiger Global Management at 1.0 to 1.06

## Analyze the Risk


```python
#Calculate the annualized standard deviation for each of the four portfolios and for the S&P 500. 
#To do that, multiply the standard deviation by the square root of the number of trading days. Use 252 for that number.

#Declare a cumulative std deviation variable by using cumulative returns and std()
cumulative_std = cumulative_returns.std()

#Declare annualized std variable by multipling cumulative_std by sqrt(252)
annualized_std = cumulative_std * math.sqrt(252)

#can set cumulative_returns.rolling(window=21).mean().plot() as a variable instead
display(cumulative_returns.rolling(window=21).mean().plot(title="Cumulative Returns w/ 21d Rolling"))
display(fund_performance.rolling(window=21).mean().plot(title="Fund Performance w/ 21d Rolling"))
display(annualized_std)

#annualized_std.rolling(window=21).mean().plot()
#annualized_std.rolling(window=21).mean()
```

## Analyze the Risk/Return Profile

```python
#Risk-Return Profile
annualized_avg_return = whale_navs_df/252
#annualized_avg_return.head()

#Create sharpe_ratio variable
sharpe_ratio = annualized_avg_return/annualized_std

sharpe_ratio

sharpe_ratio.plot(title="Sharpe Ratios", figsize=(20,12))
```

## Diversify the Portfolio

```python
#Variance

#Create 60-day rolling variables for S&P, Berkshire and Tiger Global Management
SP_60d = cumulative_returns["S&P 500"].rolling(window=60)
Berk_60d = cumulative_returns["BERKSHIRE HATHAWAY INC"].rolling(window=60)
Tiger_60d = cumulative_returns["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=60)

#Create Variance variables
SP_var = SP_60d.var()
Berk_var = Berk_60d.var()
Tiger_var = Tiger_60d.var()

#display(SP_var)
#display(Berk_var)
#display(Tiger_var)
```

```python
#Berkshire Hathaway Covariance and Beta Plot

#create variables
Berk_cov = Berk_60d.cov(cumulative_returns["S&P 500"])
Berk_beta = Berk_cov/Berk_var

display(Berk_beta.mean())
display(Berk_beta.plot(title="Berkshire Hathaway Beta"))
```

```python
#Tiger Global Covariance and Beta Plot

#create variables
Tiger_cov = Tiger_60d.cov(cumulative_returns["S&P 500"])
Tiger_beta = Tiger_cov/Tiger_var

display(Tiger_beta.mean())
display(Tiger_beta.plot(title="Tiger Global Mgmt Beta"))
```
